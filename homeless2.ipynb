```python
# Let's build exactly what you asked for.

import pandas as pd

# Load the two datasets
df_2024 = pd.read_csv('/content/drive/MyDrive/traffic data/2024_clien_rt_count_censusenamed.csv')
df_2018 = pd.read_csv('/content/drive/MyDrive/traffic data/pit count 2018 vs 2024/2018-Point-in-Time-Count-Census-Tract-Data.csv')

# Standardize the Census Tract column as string (important for merging)
df_2024['Census Tract'] = df_2024['Census Tract'].astype(str).str.strip()
df_2018['Census Tract'] = df_2018['Census Tract'].astype(str).str.strip()

# Rename Total Count columns for clarity
df_2024.rename(columns={'Total Count': 'Total Count 2024'}, inplace=True)
df_2018.rename(columns={'Total Count': 'Total Count 2018'}, inplace=True)

# Merge the two datasets on 'Census Tract', include all tracts (outer join)
merged_df = pd.merge(df_2024, df_2018, on='Census Tract', how='outer')

# Fill missing values with 0
merged_df['Total Count 2024'] = merged_df['Total Count 2024'].fillna(0).astype(int)
merged_df['Total Count 2018'] = merged_df['Total Count 2018'].fillna(0).astype(int)

# Fill missing Community names (optional: prefer 2024 Community, if missing take from 2018 Community)
merged_df['Community'] = merged_df['Community_x'].combine_first(merged_df['Community_y'])
merged_df.drop(columns=['Community_x', 'Community_y'], inplace=True)

# Create the Difference column
merged_df['Difference (2024 - 2018)'] = merged_df['Total Count 2024'] - merged_df['Total Count 2018']

# Save the result
output_csv_path = '/content/drive/MyDrive/traffic data/pit count 2018 vs 2024/census_tract_difference_2024_vs_2018.csv'
merged_df.to_csv(output_csv_path, index=False)


```

```python
import geopandas as gpd
import pandas as pd

# 1. Load the full California census tract shapefile
shapefile_path = "/content/drive/MyDrive/traffic data/tl_2022_06_tract.shp"
tracts = gpd.read_file(shapefile_path)

# 2. Filter to San Diego County (County FIPS code '073')
tracts_sd = tracts[tracts['COUNTYFP'] == '073'].copy()

# 3. Create GEOID if not already (normally already present)
tracts_sd['GEOID'] = tracts_sd['STATEFP'] + tracts_sd['COUNTYFP'] + tracts_sd['TRACTCE']

# 4. Load your census difference CSV
csv_path = "/content/drive/MyDrive/traffic data/pit count 2018 vs 2024/census_tract_difference_2024_vs_2018.csv"
df = pd.read_csv(csv_path)

# 5. Create matching GEOID from Census Tract
df['GEOID'] = '06073' + df['Census Tract'].astype(str).str.replace('.', '').str.zfill(6)

# 6. Merge tract geometry with your CSV data
merged = tracts_sd.merge(df, on="GEOID", how="left")

# 7. Optional: Fill missing values with 0
merged['Difference (2024 - 2018)'] = merged['Difference (2024 - 2018)'].fillna(0)

# 8. Export as GeoJSON
output_geojson = "/content/drive/MyDrive/traffic data/pit count 2018 vs 2024/census_difference_sandiego.geojson"
merged.to_file(output_geojson, driver="GeoJSON")

print("✅ Saved:", output_geojson)

```
